```python
import matplotlib.pyplot as plt   # Отрисовка изображений в Colab
from PIL import Image, ImageDraw  # Библиотека для работы с изображениями
import math
import numpy as np
import sys
import os

from google.colab import drive
drive.mount('/content/drive')
```

```python
def showImage(img):
  img = img.astype('uint8')           # Меняем тип на однобайтовый без знака
  img = Image.fromarray(img)          # Переводим массив в изображение
  img = img.convert('RGBA')           # Конвертируем в нужное цветовое представление
  plt.imshow(img)                     # Отрисовываем картинку
  plt.show()                          # Выводим картинку
  
```

Загрузка изображения

```python
catFile = '/content/drive/MyDrive/Colab Notebooks/Machine_learning/Control_points/CP1_Add_sth_to_image/Task1/Кот.jpg'

# Сохраним изображение кота в сущ.
cat_img = Image.open(catFile)

# Преобразумем к формату RGBA (с каналом прозрачности)
cat_img = cat_img.convert('RGBA')

# Отобразим его
plt.imshow(cat_img)
plt.show()
```

Изменение размера изображения

```python
#Привести размер изображения к 300*300 

# Вычисление пропорций для изменения размера
width, height = cat_img.size # текущие ширина и высота
new_height = 300 # новая высота
new_width = round(new_height / height * width) # новая ширина
resized_p = cat_img.resize((new_width, new_height), Image.ANTIALIAS)

# Отобразим оригинальное изображение
plt.imshow(cat_img)
plt.show()

# Отобразим изображение после пропрционально обрезки
plt.imshow(resized_p)
plt.show()
```

Обрезка изображения для итогового приведения размера

```python
# Функция crop 
# x y - верхняя левая точка рамки обрезки
# x y - нижняя правая точка рамки обрезки

cropped = resized_p.crop((0, 0, 300, 300))

# Отобразим его
plt.imshow(cropped)
plt.show()
```

Нарисовать оранжевый квадрат 20*20

```python
img_data = cropped.getdata()  # Получаем данные картинки
img_data = np.array(img_data) # Преобразуем в np массив

img      = img_data.reshape(300, 300, 4)  # И решейпим изображение в нужную  (Массив 300 на 300 пикселей и 3 - кол. цветов (RGB))
showImage(img)


```
