```python
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 1) Загрузите данные из файла по ссылке
df_data1 = pd.read_excel('http://ithub.role.ru/data/train.xlsx')
df_data1
```

```python
# Устанавливаю нужные заголовки
df_data1.columns = df_data1.iloc[0]
df_data1.drop(0, inplace=True)

cols = ["region", "code"]
cols.extend(list(df_data1)[2:])
# print(cols)
df_data1.columns = cols
df_data1
```

```python
# 2) Построить на одном поле два графика
# (не нужно делать разные сетки по оси Y)
# по перевозке пассажиров по годам по следующим регионам:

# Московская область
# Москва
```

```python
# Нужная информация для графиков
dates = cols[2:]
print(dates)
```

```python
# Строки df содержащие нужную информацию
moscow_data = df_data1.loc[df_data1['region'] == 'Город Москва столица Российской Федерации город федерального значения'].values.tolist()
moscow_region_data = df_data1.loc[df_data1['region'] == 'Московская область'].values.tolist()
```

```python
# print(moscow_data)
```

```python
moscow_data = moscow_data[0][2:]
moscow_region_data = moscow_region_data[0][2:]
```

```python
# Два графика на одном полотне
plt.title("Перевозка пассажиров из Москвы и Московской обласи")
plt.grid()              # включение отображение сетки
plt.plot(dates, moscow_data, dates, moscow_region_data)  # построение графика
```

```python
from numpy import int64
# Вывод для себя:
df_data1.code = df_data1.code.astype(int64)
df_data1
```

```python
# 3) Построить столбчатую диаграмму, где необходимо взять данные по всем областям,
# входящим в Центральный Федеральный Округ и отобразить на ней 10 лидеров по перевозкам и графу "Другие".
#  У диаграммы сделать заголовок.
```

```python
# Мои заметки:

# Строки отделяющие округа имеют значения df_data1["code"] < 1000 

# Лидеров можно определять по 2021 году

# Заметки по алгоритму

# 1. Запускаю перебор срок.
# Найду строку Северо-Западный федеральный округ
# Иду от неё вниз, пока не дойду до следующего округа. Этот момент отлавливаю используя df_data1["code"] < 1000 
```

```python
passangers_regions = dict()

inRegion = False # Не сохранять данные из областей других округов

for _, values in df_data1.iterrows():
    if values['region'] == 'Центральный федеральный округ' or inRegion: # Найден нужный округ в котором производить расчёты
        inRegion = True
        # Сбор данных о областях в этом городе
        # print(values['code'])
        if values['code'] > 1000:
            passangers_regions[values["region"]] = float(values["2021 г."])
        elif 'округ' in values['region'] and values['region'] != 'Центральный федеральный округ': # Дошли до следующего региона
            break
```

```python
passangers_regions
```

```python
passangers_regions_tuples = sorted(passangers_regions.items(), key=lambda x: x[1], reverse = True)
passangers_regions = dict(passangers_regions_tuples)
```

```python
passangers_regions_tuples
```

```python
# Разделил список кортежей на значения для столбчатой диаграммы
top_regions = [region[0] for region in passangers_regions_tuples]
top_values = [region[1] for region in passangers_regions_tuples]
print(top_regions)
print(top_values)
```

```python
# отобразить на ней 10 лидеров по перевозкам и графу "Другие".
plt.figure(figsize=(11, 6))

# Наносим надписи
plt.title("10 лидеров по перевозкам в регионе и Другие")

# Стандартная столбчатая диаграмма
regions = top_regions[:10]
regions.append("Другие")

values = top_values[:10]
values.append(sum(top_values[10:]))
plt.xticks(rotation = 90)
plt.bar(regions, values)
```

```python
# 4) Эти же данные отобразить в виде круговой диаграммы. Сделать её крупнее, чем стандартную, не используя параметр radius. У диаграммы сделать заголовок.

plt.figure(figsize=(30,15))
plt.title("10 лидеров по перевозкам в регионе и Другие")
plt.pie(values)
plt.legend(loc='best', labels=regions)

plt.show()
```
