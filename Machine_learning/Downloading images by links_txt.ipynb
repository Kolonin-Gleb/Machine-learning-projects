```python
files_path = {
    "pc_images": "/content/drive/MyDrive/Colab Notebooks/Machine_learning/pc_images_links.txt"
}
```

```python
# На случай, если скачивать изображения разных классов их разных источников
# files_path = {
#     "killerwhale": "/content/killerwhale.txt", 
#     "sharks":      "/content/sharks.txt",
#     "turtles":     "/content/turtles.txt"
# }
```

```python
# Число ссылок, что необходимо скачать из файла
links_to_download = 700

for folder, file_path in files_path.items():
    print(folder)
    links = open(file_path).read().splitlines()

    pb = ProgressBar(links_to_download)

    for i, link in enumerate(links[:links_to_download]):
        response = requests.get(link)
        img = Image.open(BytesIO(response.content))
        img.save(f"{folder}/{i}.jpg")
        pb.next()

    print()
```
