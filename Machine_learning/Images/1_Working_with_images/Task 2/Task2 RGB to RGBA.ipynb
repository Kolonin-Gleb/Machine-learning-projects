```python
# Нужно насыщенность пикселя перевести в уровень прозрачности.
# Это значит, что чем ближе код цвета к 255, тем ближе код прозрачности к 0.
def changeOpacity(image, image_height, image_width):
  image = image.reshape(image_height, image_width, 4) # Решейпим изображение в 2х массив
  
  print(image)

  # Пробегаемся по массиву и заменяем пиксели
  for h in range(image_height):
    for w in range(image_width):
      colorCode = image[h][w][0]  # Получаю значение пикселя моего изображения #0-255
      opacityCode = 6 #255 - colorCode # Определяю код прозрачности

      # Подставить массив и добавить прозрачность по 3 индексу
      image[h][w][3] = opacityCode # Добавление прозрачности
```
