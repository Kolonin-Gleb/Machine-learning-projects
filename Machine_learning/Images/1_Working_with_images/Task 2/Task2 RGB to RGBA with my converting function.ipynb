```python
def showImageAsArray(img):
  # Получаю ширину и высоту изображения
  height, width = img.size[0], img.size[1]

  img = np.array(img.getdata()) # Перевод входного изображения в массив
  img = img.reshape(height, width) # Меняем размерность массива

  # Вывод пикселей массива соответствующих пикселям изображения
  for i in range(height):
    for j in range(width):
      print(img[i][j], end='\t') #Вывод пикселей по строкам
    print('\n')

def getRGBAImageArrayFrom(img):
  # Получаю ширину и высоту изображения
  height, width = img.size[0], img.size[1]

  # Создаю 2х мерный пустой массив содержащий 4х мерные массивы для хранения цветов и прозрачности пикселей
  rgbaImageArray = np.empty((height, width, 4), dtype='uint8')

  img = np.array(img.getdata()) # Перевод входного изображения в массив
  img = img.reshape(height, width) # Меняем размерность массива

  # Заполняю пустой массив пикселями соответствующими полученному изображению и добавляю непрозрачность
  for y in range(height):
    for x in range(width):
      pixel = img[y][x]
      rgbaImageArray[y][x] = np.array([0, 0, 0, 255-pixel], dtype='uint8') # pixel - 255 = непрозрачный пискель

  return rgbaImageArray
```

```python
# Загрузка изображения

pic = '/content/drive/MyDrive/Colab Notebooks/Machine learning/Images/Working with images/Task 2/stamp.png'
stamp = Image.open(pic)

stamp = stamp.convert('L') # Приводим к режиму ЧБ

# showImageAsArray(stamp)

stampRGBAImageArray = getRGBAImageArrayFrom(stamp) # Преобразование изображения в RGBA

# showImageAsArray(stamp) # Преобразование успешно, т.к. теперь размерность массива не подходит для отображения

# Отрисовываем картинку
plt.imshow(stampRGBAImageArray)
plt.show()
```
